```python
x_axis  = np.linspace(x_min, x_max , 100)
bandwidth = ((4*data1.std()**5)/(3*len(data1)))**0.2
bandwidth
```

```python


sns.rugplot(data1);

# X軸を競ってい
x_min = data1.min() - 2
x_max = data1.max() + 2

# 全体を等間隔に100分割します。
x_axis = np.linspace(x_min,x_max,100)

bandwidth = ((4*data1.std()**5)/(3*len(data1)))**0.2

kernel_list = []

for data_point in data1:
    
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)

    kernel = kernel / kernel.max()
    kernel = kernel * 0.4
    plot.plot(x_axis,kernel,color = 'gray',alpha=0.5)

plot.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list, axis = 0)
```

```python
fig = plot.plot(x_axis, sum_of_kde, color = 'indianred')
sns.rugplot(data1)
```
